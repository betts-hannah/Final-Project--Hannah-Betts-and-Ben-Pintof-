```python
# Hannah Betts and Ben Pintof
# Final Project
# Basketball Season Simulator

# Import Packages
import pandas as pd
import numpy as np
import math as ma
import random as rand

# Names of Players
undraftedPlayersTable = pd.read_csv('all_seasons.csv')
undraftedPlayersTable = undraftedPlayersTable[['player_name', 'age', 'player_height', 'player_weight', 'college']]

# Drop duplicate rows based on 'player_name'
undraftedPlayersTable = undraftedPlayersTable.drop_duplicates(subset=['player_name'])
```

```python
# Create 'position' column based on height
def assign_position(height):
  for i in undraftedPlayersTable.index:
    pos = ''
    if height.loc[i] < 190:
      pos = 'Guard'
    elif height.loc[i] >= 190 and height.loc[i] < 208:
      pos = 'Forward'
    else:
      pos = 'Center'
    undraftedPlayersTable.loc[i, 'position'] = pos

assign_position(undraftedPlayersTable['player_height'])
```

```python
def draftingPlayers(undraftedPlayersTable, TeamsTable):
  global drafted_players_table
  for i in range(len(TeamsTable)):
    team_name = TeamsTable.loc[i, 'Team_Abbreviation']

    #Draft two Guards
    for i in range(2):
      g_player = undraftedPlayersTable[undraftedPlayersTable['position'] == 'Guard'].sample(2, replace=True) # Added replace=True for sampling with replacement if needed
      if not g_player.empty:
        player = g_player['player_name'].iloc[0] # Changed 'players' to 'player_name'
        position = g_player['position'].iloc[0]
        college = g_player['college'].iloc[0]
        age = g_player['age'].iloc[0]
        undraftedPlayersTable = undraftedPlayersTable[undraftedPlayersTable['player_name'] != player]  # Changed 'players' to 'player_name'
        new_row = pd.DataFrame({'Name': [player], 'Position': [position], 'Team_Abbreviation': [team_name], 'Age': [age],
                              'College': [college], 'Points': [0], 'Rebounds': [0], 'Assists' : [0], 'Blocks' : [0],
                              'Steals': [0], 'Games_Played': [0]})
        drafted_players_table = pd.concat([drafted_players_table, new_row], ignore_index=True)

    #Draft two Forwards
    for i in range(2):
      f_player = undraftedPlayersTable[undraftedPlayersTable['position'] == 'Forward'].sample(2, replace=True) # Added replace=True
      if not f_player.empty:
        player = f_player['player_name'].iloc[0] # Changed 'players' to 'player_name' and used f_player instead of g_player
        position = f_player['position'].iloc[0]
        college = f_player['college'].iloc[0]
        age = f_player['age'].iloc[0]
        undraftedPlayersTable = undraftedPlayersTable[undraftedPlayersTable['player_name'] != player]  # Changed 'players' to 'player_name'
        new_row = pd.DataFrame({'Name': [player], 'Position': [position], 'Team_Abbreviation': [team_name], 'Age': [age],
                              'College': [college], 'Points': [0], 'Rebounds': [0], 'Assists' : [0], 'Blocks' : [0],
                              'Steals': [0], 'Games_Played': [0]})
        drafted_players_table = pd.concat([drafted_players_table, new_row], ignore_index=True)

    #Draft one Center
    c_player = undraftedPlayersTable[undraftedPlayersTable['position'] == 'Center'].sample(1, replace=True) # Added replace=True and used c_player instead of f_player
    if not c_player.empty:
      player = c_player['player_name'].iloc[0] # Changed 'players' to 'player_name' and used c_player instead of g_player
      position = c_player['position'].iloc[0]
      college = c_player['college'].iloc[0]
      age = c_player['age'].iloc[0]
      undraftedPlayersTable = undraftedPlayersTable[undraftedPlayersTable['player_name'] != player]  # Changed 'players' to 'player_name'
      new_row = pd.DataFrame({'Name': [player], 'Position': [position], 'Team_Abbreviation': [team_name], 'Age': [age],
                              'College': [college], 'Points': [0], 'Rebounds': [0], 'Assists' : [0], 'Blocks' : [0],
                              'Steals': [0], 'Games_Played': [0]})
      drafted_players_table = pd.concat([drafted_players_table, new_row], ignore_index=True)
```

```python
def play_to_bin(PlayerTable_df):
  for i in range(len(PlayerTable_df)):
    BinsTable_df.loc[i, 'Player'] = PlayerTable_df.loc[i, 'Name']
    pnum = rand.randint(0, 10)
    anum = rand.randint(0, 10)
    rnum = rand.randint(0, 10)
    snum = rand.randint(0, 10)
    bnum = rand.randint(0, 10)

# Creating Guard Bins
    if PlayerTable_df.loc[i, 'Position'] == 'Guard':
      # Creating Point bin
      if pnum < 3:
        BinsTable_df.loc[i, 'Point_Bin'] = 'High'
      elif pnum > 7:
        BinsTable_df.loc[i, 'Point_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Point_Bin'] = 'Medium'

      # Creating Assist bin
      if anum < 4:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'High'
      elif anum > 8:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'Medium'

      # Creating Rebound bin
      if rnum < 2:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'High'
      elif rnum > 2:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'Medium'

      # Creating Steal bin
      if snum < 4:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'High'
      elif snum > 7:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'Medium'

      # Creating Block bin
      if bnum < 1:
        BinsTable_df.loc[i, 'Block_Bin'] = 'High'
      elif bnum > 1:
        BinsTable_df.loc[i, 'Block_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Block_Bin'] = 'Medium'

# Creating Forward Bins
    if PlayerTable_df.loc[i, 'Position'] == 'Forward':
      # Creating Point bin
      if pnum < 3:
        BinsTable_df.loc[i, 'Point_Bin'] = 'High'
      elif pnum > 6:
        BinsTable_df.loc[i, 'Point_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Point_Bin'] = 'Medium'

      # Creating Assist bin
      if anum < 4:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'High'
      elif anum > 8:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'Medium'

      # Creating Rebound bin
      if rnum < 2:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'High'
      elif rnum > 8:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'Medium'

      # Creating Steal bin
      if snum < 1:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'High'
      elif snum > 3:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'Medium'

      # Creating Block bin
      if bnum < 1:
        BinsTable_df.loc[i, 'Block_Bin'] = 'High'
      elif bnum > 1:
        BinsTable_df.loc[i, 'Block_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Block_Bin'] = 'Medium'

# Creating Center Bins
    if PlayerTable_df.loc[i, 'Position'] == 'Center':
      # Creating Point bin
      if pnum < 2:
        BinsTable_df.loc[i, 'Point_Bin'] = 'High'
      elif pnum > 8:
        BinsTable_df.loc[i, 'Point_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Point_Bin'] = 'Medium'

      # Creating Assist bin
      if anum < 1:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'High'
      elif anum > 3:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Assist_Bin'] = 'Medium'

      # Creating Rebound bin
      if rnum < 4:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'High'
      elif rnum > 8:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Rebound_Bin'] = 'Medium'

      # Creating Steal bin
      if snum > 0:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Steal_Bin'] = 'Medium'

      # Creating Block bin
      if bnum < 3:
        BinsTable_df.loc[i, 'Block_Bin'] = 'High'
      elif bnum > 5:
        BinsTable_df.loc[i, 'Block_Bin'] = 'Low'
      else:
        BinsTable_df.loc[i, 'Block_Bin'] = 'Medium'
```

```python
# Creating Player Stats Function using the Bins
def player_stats(Player, Points, Assists, Rebounds, Steals, Blocks):
  pointbin = BinsTable_df.loc[Player, 'Point_Bin']
  assistbin = BinsTable_df.loc[Player, 'Assist_Bin']
  reboundbin = BinsTable_df.loc[Player, 'Rebound_Bin']
  stealbin = BinsTable_df.loc[Player, 'Steal_Bin']
  blockbin = BinsTable_df.loc[Player, 'Block_Bin']

# Finding the Number of Points scored by a player
  tempnum = rand.randint(0, 30)
  if pointbin == 'High':
    if tempnum == 19:
      Points = rand.randint(2, 18)
    elif tempnum == 14:
      Points = rand.randint(30, 70)
    else:
      Points = rand.randint(19, 29)
  elif pointbin == 'Medium':
    if tempnum == 19:
      Points = rand.randint(0, 9)
    elif tempnum == 14:
      Points = rand.randint(19, 32)
    else:
      Points = rand.randint(10, 18)
  else:
    if tempnum == 19:
      Points = rand.randint(0, 4)
    elif tempnum == 14:
      Points = rand.randint(10, 19)
    else:
      Points = rand.randint(5, 9)


# Finding the Number of Assists by a Player
  tempnum = rand.randint(0, 20)
  if assistbin == 'High':
    if tempnum == 19:
      Assists = rand.randint(0, 5)
    elif tempnum == 14:
      Assists = rand.randint(8, 15)
    else:
      Assists = rand.randint(5, 7)
  elif assistbin == 'Medium':
    if tempnum == 19:
      Assists = rand.randint(0, 2)
    elif tempnum == 14:
      Assists = rand.randint(6, 10)
    else:
      Assists = rand.randint(3, 5)
  else:
    if tempnum == 19:
      Assists = 0
    elif tempnum == 14:
      Assists = rand.randint(3, 6)
    else:
      Assists = rand.randint(0, 2)

# Finding the Number of Rebounds by a Player
  tempnum = rand.randint(0, 20)
  if reboundbin == 'High':
    if tempnum == 19:
      Rebounds = rand.randint(0, 5)
    elif tempnum == 14:
      Rebounds = rand.randint(8, 15)
    else:
      Rebounds = rand.randint(5, 7)
  elif reboundbin == 'Medium':
    if tempnum == 19:
      Rebounds = rand.randint(0, 2)
    elif tempnum == 14:
      Rebounds = rand.randint(6, 10)
    else:
      Rebounds = rand.randint(3, 5)
  else:
    if tempnum == 19:
      Rebounds = 0
    elif tempnum == 14:
      Rebounds = rand.randint(3, 6)
    else:
      Rebounds = rand.randint(0, 2)


# Finding the Number of Steals by a Player
  tempnum = rand.randint(0, 20)
  if stealbin == 'High':
    if tempnum == 19:
      Steals = 0
    elif tempnum == 14:
      Steals = rand.randint(1, 6)
    else:
      Steals = rand.randint(0, 3)
  elif stealbin == 'Medium':
    if tempnum == 19:
      Steals = 0
    elif tempnum == 14:
      Steals = rand.randint(1, 3)
    else:
      Steals = rand.randint(0, 1)
  else:
    if tempnum == 19:
      Steals = 0
    elif tempnum == 14:
      Steals = rand.randint(0, 2)
    else:
      Steals = 0

# Finding the Number of Blocks by a Player
  tempnum = rand.randint(0, 20)
  if blockbin == 'High':
    if tempnum == 19:
      Blocks = 0
    elif tempnum == 14:
      Blocks = rand.randint(1, 6)
    else:
      Blocks = rand.randint(0, 3)
  elif blockbin == 'Medium':
    if tempnum == 19:
      Blocks = 0
    elif tempnum == 14:
      Blocks = rand.randint(1, 3)
    else:
      Blocks = rand.randint(0, 1)
  else:
    if tempnum == 19:
      Blocks = 0
    elif tempnum == 14:
      Blocks = rand.randint(0, 2)
    else:
      Blocks = 0


  return Points, Assists, Rebounds, Steals, Blocks
```

```python
# Creating a function that makes each team play each other
def PlayEachOther():
  global game_log_player_table, game_log_team_table
  gID = 1
  for k in range(2):
    for i in range(len(TeamsTable['Team_Abbreviation'])):
      for j in range(i + 1, len(TeamsTable['Team_Abbreviation'])):
        Team1Points = 0
        Team1Assists = 0
        Team1Rebounds = 0
        Team1Steals = 0
        Team1Blocks = 0
        Team2Points = 0
        Team2Assists = 0
        Team2Rebounds = 0
        Team2Steals = 0
        Team2Blocks = 0
        currentTeam1 = TeamsTable.loc[i, 'Team_Abbreviation']
        currentTeam2 = TeamsTable.loc[j, 'Team_Abbreviation']
        for p in range(len(drafted_players_table['Name'])):
          if drafted_players_table.loc[p, 'Team_Abbreviation'] == currentTeam1 or drafted_players_table.loc[p, 'Team_Abbreviation'] == currentTeam2:
            if drafted_players_table.loc[p, 'Team_Abbreviation'] == currentTeam1:
              (playPoints, playAssists, playRebounds, playSteals, playBlocks) = player_stats(drafted_players_table.loc[p, 'Name'], 0, 0, 0, 0, 0)
              Team1Points += playPoints
              Team1Assists += playAssists
              Team1Rebounds += playRebounds
              Team1Steals += playSteals
              Team1Blocks += playBlocks
            elif drafted_players_table.loc[p, 'Team_Abbreviation'] == currentTeam2:
              (playPoints, playAssists, playRebounds, playSteals, playBlocks) = player_stats(drafted_players_table.loc[p, 'Name'], 0, 0, 0, 0, 0)
              Team2Points += playPoints
              Team2Assists += playAssists
              Team2Rebounds += playRebounds
              Team2Steals += playSteals
              Team2Blocks += playBlocks
            drafted_players_table.loc[p, 'Points'] += playPoints
            drafted_players_table.loc[p, 'Assists'] += playAssists
            drafted_players_table.loc[p, 'Rebounds'] += playRebounds
            drafted_players_table.loc[p, 'Steals'] += playSteals
            drafted_players_table.loc[p, 'Blocks'] += playBlocks
            drafted_players_table.loc[p, 'Games_Played'] += 1
            name = drafted_players_table.loc[p, 'Name']
            new_row = pd.DataFrame({'GameID' : [gID], 'Name': [name], 'Points' : [playPoints], 'Rebounds' : [playRebounds],
                                      'Assists' : [playAssists], 'Blocks' : [playSteals], 'Steals' : [playSteals]})
            game_log_player_table = pd.concat([game_log_player_table, new_row], ignore_index=True)
        if Team1Points > Team2Points:
          res1 = 'W'
          res2 = 'L'
          TeamsTable.loc[i, 'Wins'] += 1
          TeamsTable.loc[j, 'Losses'] += 1
        elif Team1Points < Team2Points:
          res1 = 'L'
          res2 = 'W'
          TeamsTable.loc[i, 'Losses'] += 1
          TeamsTable.loc[j, 'Wins'] += 1
        elif Team1Points == Team2Points:
          x = rand.randint(0, 1)
          if x == 0:
            TeamsTable.loc[i, 'Wins'] += 1
            TeamsTable.loc[j, 'Losses'] += 1
            res1 = 'W'
            res2 = 'L'
          else:
            TeamsTable.loc[i, 'Losses'] += 1
            TeamsTable.loc[j, 'Wins'] += 1
            res1 = 'L'
            res2 = 'W'
        TeamsTable.loc[i, 'GamesPlayed'] += 1
        TeamsTable.loc[j, 'GamesPlayed'] += 1
        new1_row = pd.DataFrame({'GameID' : [gID], 'Team': [currentTeam1], 'Result' : [res1], 'Points' : [Team1Points],
                                 'Points Allowed' : [Team2Points], 'Assists' : [Team1Assists], 'Rebounds' : [Team1Rebounds],
                                 'Blocks' : [playSteals], 'Steals' : [playSteals]})
        game_log_team_table = pd.concat([game_log_team_table, new1_row], ignore_index=True)
        new2_row = pd.DataFrame({'GameID' : [gID], 'Team': [currentTeam2], 'Result' : [res2], 'Points' : [Team2Points],
                                 'Points Allowed' : [Team1Points], 'Assists' : [Team2Assists], 'Rebounds' : [Team2Rebounds],
                                 'Blocks' : [playSteals], 'Steals' : [playSteals]})
        game_log_team_table = pd.concat([game_log_team_table, new2_row], ignore_index=True)
        gID += 1
```
