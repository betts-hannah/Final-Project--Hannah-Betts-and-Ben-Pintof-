```python
# Hannah Betts and Ben Pintof
# Final Project
# Basketball Season Simulator

# Import Packages
import pandas as pd
import numpy as np
import math as ma

# Names of Players
undraftedPlayersTable = pd.read_csv('/all_seasons.csv')
undraftedPlayersTable = undraftedPlayersTable[['player_name', 'age', 'player_height', 'player_weight', 'college']]

# Drop duplicate rows based on 'player_name'
undraftedPlayersTable = undraftedPlayersTable.drop_duplicates(subset=['player_name'])
```

```python
# Create 'position' column based on height
def assign_position(height):
  for i in undraftedPlayersTable.index:
    pos = ''
    if height.loc[i] < 190:
      pos = 'Guard'
    elif height.loc[i] >= 190 and height.loc[i] < 208:
      pos = 'Forward'
    else:
      pos = 'Center'
    undraftedPlayersTable.loc[i, 'position'] = pos

assign_position(undraftedPlayersTable['player_height'])
```
