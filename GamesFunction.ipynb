```python
# Trying to create a basic simulation
# Creating a wins function
import pandas as pd
import math as ma
import random as rand

# creating 4 teams
TeamsTable = {
    'Teams' : ['Team A', 'Team B', 'Team C', 'Team D'],
    'Wins' : [0, 0, 0, 0],
    'Draws' : [0, 0, 0, 0],
    'Losses' : [0, 0, 0, 0],
    'Goals_For' : [0, 0, 0, 0],
    'Goals_Against' : [0, 0, 0, 0],
    'Games_Played' : [0, 0, 0, 0]
}
TeamsTable = pd.DataFrame(TeamsTable)
```

```python
# Creating a function that makes each team play each other
def PlayEachOther():
  for i in range(len(TeamsTable['Teams'])):
    for j in range(i + 1, len(TeamsTable['Teams'])):
      # Simulate a match between team i and team j
      team1_goals = rand.randint(0, 5)
      team2_goals = rand.randint(0, 5)

      # Update the table based on the match result
      if team1_goals > team2_goals:
          TeamsTable.loc[i, 'Wins'] += 1
          TeamsTable.loc[j, 'Losses'] += 1
      elif team1_goals < team2_goals:
          TeamsTable.loc[i, 'Losses'] += 1
          TeamsTable.loc[j, 'Wins'] += 1
      else:
          TeamsTable.loc[i, 'Draws'] += 1
          TeamsTable.loc[j, 'Draws'] += 1

      TeamsTable.loc[i, 'Goals_For'] += team1_goals
      TeamsTable.loc[i, 'Goals_Against'] += team2_goals
      TeamsTable.loc[i, 'Games_Played'] += 1

      TeamsTable.loc[j, 'Goals_For'] += team2_goals
      TeamsTable.loc[j, 'Goals_Against'] += team1_goals
      TeamsTable.loc[j, 'Games_Played'] += 1
```

```python
PlayEachOther()
```
