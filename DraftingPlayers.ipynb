```python
# We need to create a function to draft NBA players onto teams
import pandas as pd
import math as ma
import numpy as np
import random as rand

undraftedPlayersTable = {
    'players' : ['playerA', 'playerB', 'playerC', 'playerD', 'playerE', 'playerF', 'playerG', 'playerH'],
    'position' : ['G', 'G', 'F', 'F', 'G', 'G', 'F', 'F']
}

undraftedPlayersTable = pd.DataFrame(undraftedPlayersTable)

draftedTable = {
    'players' : [],
    'position' : [],
    'team' : []
}

draftedTable = pd.DataFrame(draftedTable)

TeamsTable = {
    'Teams' : ['TeamA', 'TeamB', 'TeamC'],
    'Location' : ['CityA', 'CityB', 'CityC']
}

TeamsTable = pd.DataFrame(TeamsTable)

teamA = {
    'players' : [],
    'position' : []
}

teamA = pd.DataFrame(teamA)

teamB = {
    'players' : [],
    'position' : []
}

teamB = pd.DataFrame(teamB)

teamC = {
    'players' : [],
    'position' : []
}

teamC = pd.DataFrame(teamC)

def draftingPlayers(undraftedPlayersTable, TeamsTable):
  global draftedTable
  for i in range(len(TeamsTable)):
    team_name = TeamsTable.loc[i, 'Teams']
    #Draft one G
    g_player = undraftedPlayersTable[undraftedPlayersTable['position'] == 'G'].sample(1)
    if not g_player.empty:
      player = g_player['players'].iloc[0]
      position = g_player['position'].iloc[0]
      undraftedPlayersTable = undraftedPlayersTable[undraftedPlayersTable['players'] != player]
      new_row = pd.DataFrame({'players': [player], 'position': [position], 'team': [team_name]})
      draftedTable = pd.concat([draftedTable, new_row], ignore_index=True)

    #Draft one F
    f_player = undraftedPlayersTable[undraftedPlayersTable['position'] == 'F'].sample(1)
    if not f_player.empty:
      player = f_player['players'].iloc[0]
      position = f_player['position'].iloc[0]
      undraftedPlayersTable = undraftedPlayersTable[undraftedPlayersTable['players'] != player]
      new_row = pd.DataFrame({'players': [player], 'position': [position], 'team': [team_name]})
      draftedTable = pd.concat([draftedTable, new_row], ignore_index=True)
```

```python
draftingPlayers(undraftedPlayersTable, TeamsTable)
```
